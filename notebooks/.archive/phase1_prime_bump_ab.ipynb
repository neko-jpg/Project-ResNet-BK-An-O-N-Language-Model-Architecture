# Phase1 prime-bump A/B (Colab ready)
- Clone repo (if needed)
- Install deps
- Train baseline (no prime-bump)
- Train prime-bump (scale 0.02)
- Logs: `checkpoints/logs` (loss/PPL/NaN有無を比較)


```python
import subprocess, pathlib
repo_url = "https://github.com/neko-jpg/Project-ResNet-BK-An-O-N-Language-Model-Architecture.git"
cwd = pathlib.Path('.').resolve()
if (cwd / '.git').exists():
    repo_root = cwd
else:
    repo_root = cwd / 'Project-ResNet-BK-An-O-N-Language-Model-Architecture'
    if not repo_root.exists():
        print(f'Cloning {repo_url} into {repo_root} ...')
        subprocess.run(['git', 'clone', repo_url, str(repo_root)], check=True)
    else:
        print(f'Using existing clone at {repo_root}')
print('Repo root:', repo_root)
```

```python
import subprocess, sys
print('Installing requirements...')
res = subprocess.run([sys.executable, '-m', 'pip', 'install', '-r', 'requirements.txt'], cwd=repo_root, check=False)
print('returncode', res.returncode)
assert res.returncode == 0, 'pip install failed'
```

```python
import subprocess, sys
cmd = [sys.executable, 'train.py', '--config-preset', 'baseline']
print('Running (no prime-bump):', ' '.join(cmd))
res = subprocess.run(cmd, cwd=repo_root, check=False)
print('returncode', res.returncode)
assert res.returncode == 0, 'train.py no-prime failed'
print('Logs in checkpoints/logs')
```

```python
import subprocess, sys
cmd = [sys.executable, 'train.py', '--config-preset', 'baseline', '--prime-bump-init', '--prime-bump-scale', '0.02']
print('Running (prime-bump):', ' '.join(cmd))
res = subprocess.run(cmd, cwd=repo_root, check=False)
print('returncode', res.returncode)
assert res.returncode == 0, 'train.py prime-bump failed'
print('Logs in checkpoints/logs')
```
